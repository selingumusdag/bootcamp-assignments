```python
import pandas as pd
import numpy as np
titanic=pd.read_csv('./data/train.csv')
titanic.head()
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
titanic.Age = titanic.Age.fillna(titanic.Age.mean())
```

```python
titanic['Age']=titanic['Age'].astype(int)
```

```python
titanic.Embarked = titanic.Embarked.fillna(titanic.Embarked.mode())
```

```python
titanic.head()
```

```python
titanic=titanic.drop(columns=['Name'])
titanic=titanic.drop(columns=['Cabin'])
titanic=titanic.drop(columns=['Ticket'])
titanic=titanic.drop(columns=['PassengerId'])
```

```python
titanic.head()
```

```python
titanic_sex = pd.get_dummies(titanic['Sex'])
titanic_embarked = pd.get_dummies(titanic['Embarked'])
```

```python
titanic = pd.concat([titanic, titanic_sex,titanic_embarked], axis=1)
titanic.head()
```

```python
titanic=titanic.drop(columns=['Sex'])
titanic=titanic.drop(columns=['Embarked'])
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
Y = titanic["Survived"]
X = titanic[["Pclass","Age","SibSp","Parch","Fare","female","male","C","Q","S"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, Y, test_size=0.20, random_state=111)
```

```python
log_reg.fit(X_train,y_train)
```

```python
train_accuracy = log_reg.score(X_train, y_train)
test_accuracy = log_reg.score(X_test, y_test)

print('One-vs.-Rest', '-'*20, 
      'Accuracy on Train Data : {:.2f}'.format(train_accuracy), 
      'Accuracy on Test Data  : {:.2f}'.format(test_accuracy), sep='\n')
```

```python
log_reg_mnm = LogisticRegression(multi_class='multinomial', solver='lbfgs')
log_reg_mnm.fit(X_train, y_train)

train_accuracy = log_reg_mnm.score(X_train, y_train)
test_accuracy = log_reg_mnm.score(X_test, y_test)

print('Multinomial (Softmax)', '-'*20, 
      'Accuracy on Train Data : {:.2f}'.format(train_accuracy), 
      'Accuracy on Test Data  : {:.2f}'.format(test_accuracy), sep='\n')
```

```python
#predictions = log_reg.predict(X_test)
C_values = [0.001,0.01, 0.1,1,10,100, 1000]

accuracy_values = pd.DataFrame(columns=['C_values', 'Train Accuracy', 'Test Accuracy'])

for c in C_values:
    # Apply logistic regression model to training data
    lr = LogisticRegression(penalty = 'l2', C = c, random_state = 0, solver='lbfgs', multi_class='multinomial')
    lr.fit(X_train, y_train)
    accuracy_values = accuracy_values.append({'C_values': c,
                                              'Train Accuracy': lr.score(X_train, y_train),
                                              'Test Accuracy': lr.score(X_test, y_test)
                                             }, ignore_index=True)
display(accuracy_values)
```
